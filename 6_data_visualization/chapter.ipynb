### Adjust Spacing around Subplots

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color='black', alpha=0.5)

fig.subplots_adjust(wspace=0, hspace=0)
```

### Ticks, Labels, Legends

```python
fig, ax = plt.subplots()

ax.plot(np.random.standard_normal(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=30, fontsize=8)

ax.set_xlabel("Stages")
ax.set_title("Matplotlib plot")
```

```python
fig, ax = plt.subplots()

ax.plot(np.random.standard_normal(1000).cumsum(), color='black', label='one')
ax.plot(np.random.standard_normal(1000).cumsum(), color='black', alpha=0.8, label='two', linestyle='dashed')
ax.plot(np.random.standard_normal(1000).cumsum(), color='black', alpha=0.6, label='three', linestyle='dotted')
plt.legend()
```

```python
from datetime import datetime
import pandas as pd

fig, ax = plt.subplots()

location = r"C:\Users\William Costa\Documents\repositories\data_science_basic_knowledge\resources\spx.csv"
data = pd.read_csv(location, index_col=0, parse_dates=True)
spx = data["SPX"]

spx.plot(ax=ax, color='black')

crisis_data = [
    (datetime(2007,10,11), 'Peak of bull market'),
    (datetime(2008,3,12), 'Bear Stearns Fails'),
    (datetime(2009,9,15), 'Leham Bankruptcy'),
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                    xytext=(date, spx.asof(date)+ 225),
                    arrowprops=dict(
                        facecolor='green', headwidth=4, width=2, headlength=2
                    ),
                    horizontalalignment="left", verticalalignment='top'
                )
    
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title("Important dates in the 2008-2009 financial crisis")
```

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[.15, .15],[.35,.4],[.2, .6]], color='green', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
location = r"C:\Users\William Costa\Documents\repositories\data_science_basic_knowledge\resources\tips.csv"
tips = pd.read_csv(location)

party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts = party_counts.reindex(index=['Thur','Fri','Sat', 'Sun'])
party_counts
```

```python
party_counts = party_counts.iloc[:, 1:5]
party_counts
```

```python
party_pct = party_counts.div(party_counts.sum(axis='columns'), axis='index')
party_pct
```

```python
party_pct.plot.bar(stacked=True)
```
